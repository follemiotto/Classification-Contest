```python
df=pd.read_csv(fpath+'adult.data')
df.head()
#df.describe()
```

```python
print('How many NaN values by attribute:')
print(df.isnull().sum())
print('How many 0 values by attribute:')
print((df==0).sum())
print('How many ? values by attribute:')
print((df=='?').sum())
```

```python
testdf=pd.read_csv(fpath+'adult.test')
testdf.head()
#testdf.describe()
```

```python
print('How many NaN values by attribute:')
print(testdf.isnull().sum())
print('How many 0 values by attribute:')
print((testdf==0).sum())
print('How many ? values by attribute:')
print((testdf=='?').sum())
```

```python
df=df[df['workclass']!='?']
testdf=testdf[testdf['workclass']!='?']
df=df[df['occupation']!='?']
testdf=testdf[testdf['occupation']!='?']
df=df[df['native-country']!='?']
testdf=testdf[testdf['native-country']!='?']
```

```python
testdf.head()
```

```python
df.drop(['education'],axis=1,inplace=True)
testdf.drop(['education'],axis=1,inplace=True)

df.head()
```

```python
laben=pp.LabelEncoder()

for feature in df.columns:
  if df[feature].dtype=='O':
    laben.fit(df[feature])
    df[feature]=laben.transform(df[feature])
    testdf[feature]=laben.transform(testdf[feature])
```

#trying every Model

```python
X_tr=df.drop(['salary50k'],axis=1,inplace=False)
y_tr=df['salary50k'].astype(int)
X_te=testdf.drop(['salary50k'],axis=1,inplace=False)
y_te=testdf['salary50k'].astype(int)
```

```python
myrf=rf()
myrf.fit(X_tr,y_tr)
y_hat=myrf.predict(X_te)
print('Raw Accuracy %.2f  Recall %.2f  Precision %.2f  F1-score %.2f'%(acc(y_hat,y_te),rec(y_hat,y_te),prec(y_hat,y_te),f1(y_hat,y_te,average='micro')))
```

```python
myet=et()
myet.fit(X_tr,y_tr)
y_hat=myet.predict(X_te)
print('Raw Accuracy %.2f  Recall %.2f  Precision %.2f  F1-score %.2f'%(acc(y_hat,y_te),rec(y_hat,y_te),prec(y_hat,y_te),f1(y_hat,y_te,average='micro')))
```

```python
myab=ab()
myab.fit(X_tr,y_tr)
y_hat=myab.predict(X_te)
print('Raw Accuracy %.2f  Recall %.2f  Precision %.2f  F1-score %.2f'%(acc(y_hat,y_te),rec(y_hat,y_te),prec(y_hat,y_te),f1(y_hat,y_te,average='micro')))
```

```python
mybg=bg()
mybg.fit(X_tr,y_tr)
y_hat=mybg.predict(X_te)
print('Raw Accuracy %.2f  Recall %.2f  Precision %.2f  F1-score %.2f'%(acc(y_hat,y_te),rec(y_hat,y_te),prec(y_hat,y_te),f1(y_hat,y_te,average='micro')))
```

```python
mygb=gb()
mygb.fit(X_tr,y_tr)
y_hat=mygb.predict(X_te)
print('Raw Accuracy %.2f  Recall %.2f  Precision %.2f  F1-score %.2f'%(acc(y_hat,y_te),rec(y_hat,y_te),prec(y_hat,y_te),f1(y_hat,y_te,average='micro')))
```

```python
myst=st(estimators=[('rf', rf(n_estimators=10, random_state=42)),('svr', mp(pp.StandardScaler(),LinearSVC(random_state=42)))])
myst.fit(X_tr,y_tr)
y_hat=myst.predict(X_te)
print('Raw Accuracy %.2f  Recall %.2f  Precision %.2f  F1-score %.2f'%(acc(y_hat,y_te),rec(y_hat,y_te),prec(y_hat,y_te),f1(y_hat,y_te,average='micro')))
```
